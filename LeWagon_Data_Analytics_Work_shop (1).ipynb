#We could drop some columns that are not relevant for us

```python
columns_to_drop = ['id', 'host_name', 'last_review']
listings_df.drop(columns_to_drop, axis='columns', inplace=True) #inplace True changes the original data frame 
```

```python
listings_df.head() #to see the change
```

Let's Remove 'NaN' from the 'reviews_per_mpnth' column

```python
listings_df.fillna({'reviews_per_month': 0}, inplace=True)
```

# Using seaborn

```python
sns.countplot(data=listings_df, x='neighbourhood_group')
```

### Distributions

```python
sns.histplot(listings_df['price']) # not so readable we would need to exlude really expensive listings
```

```python
listings_df[listings_df['price'] <= 500].shape
```

```python
affordable_df = listings_df[listings_df['price'] <= 500]
```

```python
sns.histplot(affordable_df['price'])
```

```python
affordable_df['price'].mean()
```

## What is the distribution of flat prices based on neighbourhood group

```python
sns.violinplot(data = affordable_df, x='neighbourhood_group', y='price')
```
